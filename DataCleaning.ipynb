```python
#list for activities for times
duration = data['duration'].tolist()
duration
```

```python
newduration =[]
for datapoint in duration:
    if datapoint<10.00:
        newduration.append(datapoint*60)
    if datapoint>10.00:
        newduration.append(datapoint)
newduration
data['duration']= newduration
```

```python
date = data['date'].tolist()
date
```

```python
alldates =[]
for datum in date:
    dat = str(datum)
    number = int(dat[0]) +int(dat[1])
    if number>2 or dat[0]+dat[1]=='10' or dat[0]+dat[1]=='11':
        day= dat[3:5] + '/' + dat[0:3] + dat[6:]
        alldates.append(day)
    else:
        day=dat
        alldates.append(day)
```

```python
data['date']= alldates
data.to_csv('datapart1.activities.csv')
```
